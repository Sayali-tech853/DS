```python
df['Age']=df['Age'].fillna(np.mean(df['Age']))
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

### 1.Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information 
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to 
see if we can find any patterns in the data

### Histplot

```python
# use column age and fare

sns.histplot(x=df['Age'])
```

```python
sns.histplot(x=df['Fare'])
```

### Scatter plot

```python
sns.jointplot(x=df['Age'],y=df['Fare'],kind='scatter')
```

### Hexplot

```python
sns.jointplot(x=df['Age'],y=df['Fare'],kind='hex')
```

### Rug plot

```python
sns.rugplot(df['Fare']) #small bars along the x axis for each point in the dataset 
```

### Bar plot

```python
sns.barplot(x='Sex',y='Age',data=df)
```

### Countplot

```python
sns.countplot(x=df['Sex'])
```

### Boxplot

```python
sns.boxplot(x=df['Sex'],y=df['Age'])
```

### Violin Plot

```python
sns.violinplot(x=df['Sex'],y=df['Age'])  #same as box plot first parameter is the categorical column and second parameter is the numerical column
```

### Advance Plot - Stripplot

```python
sns.stripplot(x=df['Sex'],y=df['Age']) #same as scater plot first parameter is the categorical column and second parameter is the numerical column
```

### Swarm Plot - combination of strip and violin plot

```python
sns.swarmplot(x=df['Sex'],y=df['Age']) # In this plots datapoints are not overlappedd like violin plot 
```

### Matrix Plot - Heat maps,Cluster map

```python
sns.distplot(df['Age'])
```

```python
sns.jointplot(x = "Survived", y = "Fare", kind = "scatter", data = df)
```

```python
tc = df.corr()
sns.heatmap(tc, cmap="YlGnBu")
plt.title('Correlation')
```

```python
sns.catplot(x='Pclass', y='Fare', data=df, kind='bar')
```

### 3 D plot

```python
x_column = 'Age'
y_column = 'Fare'
z_column = 'Pclass'

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[x_column], df[y_column], df[z_column], c='blue', marker='o', alpha=0.5)

ax.set_xlabel(x_column)
ax.set_ylabel(y_column)
ax.set_zlabel(z_column)

ax.set_title('3D Scatter Plot for Titanic Data')

plt.show()
```
